```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df5 = pd.read_csv('df5.csv')
print(df5.shape)
df5.head()
```

#### For making a model we should only have numerical features.
#### In this dataset, 'location' is not numerical, so we will convert it into numerical.

# One hot Encoding

```python
dummies = pd.get_dummies(df5.location)
dummies.head(3)
```

```python
df6 = pd.concat([df5, dummies.drop('other', axis='columns')], axis='columns')
df6.head(3)
```

### Now we will drop the 'location' feature

```python
df7 = df6.drop('location', axis='columns')
df7.head(3)
```

# Divide the Dataset

##### Now, we will divide the dataset into dependent and target feature

```python
X = df7.drop('price', axis='columns')
X.head(3)
```

```python
y = df7.price
y.head(3)
```

# Model Making

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=10)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
linear_model.score(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_model = DecisionTreeRegressor()
tree_model.fit(X_train, y_train)
tree_model.score(X_test, y_test)
```
