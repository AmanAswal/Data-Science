# Feature Engineering

We will be performing all the below steps in Feature Engineering:
    1. Missing values
    2. Temporal values
    3. Categorical variables: remove rare labels
    4. Standardise the values of variables to the same range

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#to visualise all the columns of the dataframe
pd.pandas.set_option('display.max_columns',None)
```

```python
df=pd.read_csv('train.csv')
df.head()
```

## Misssing Values

```python
# In categorical features
features_nan=[feature for feature in df.columns if df[feature].isnull().sum() > 1 and df[feature].dtypes == 'O']

for feature in features_nan:
    print('{}: {}% missing values'.format(feature,np.round(df[feature].isnull().mean(),4)))
```

```python
## replace the missing value with a new label

def replace_cat_features(df, features_nan):
    data=df.copy()
    data[features_nan] = data[features_nan].fillna('Missing')
    return data

df= replace_cat_features(df, features_nan)
df[features_nan].isnull().sum()
```

```python
df.head()
```

```python
# In Numerical features
numerical_with_nan = [feature for feature in df.columns if df[feature].isnull().sum() > 1 and df[feature].dtypes != 'O']

for feature in numerical_with_nan:
    print('{}: {}% missing values'.format(feature, np.round(df[feature].isnull().mean(),4)))
```

```python
# replacing the NaN values
for feature in numerical_with_nan:
    median_value = df[feature].median()
    
    df[feature + 'nan'] = np.where(df[feature].isnull(),1,0)
    df[feature].fillna(median_value,inplace = True)

df[numerical_with_nan].isnull().sum()
                                                               
        
```

```python
df.head()
```

```python
num_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    df[feature]= np.log(df[feature])
```

```python
df.head()
```

## Handling Rare Categorical features
We will remove categorical variables that are present less than 1% of the observation

```python
categorical_features = [feature for feature in df.columns if df[feature].dtypes == 'O']
categorical_features
```

```python
for feature in categorical_features:
    temp = df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df = temp[temp>0.01].index
    df[feature] = np.where(df[feature].isin(temp_df),df[feature], 'Rare_var')
```

```python
df.head(50)
```

```python
for feature in categorical_features:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```

## Feature Scaling

```python
feature_scale = [feature for feature in df.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df[feature_scale])
```

```python
scaler.transform(df[feature_scale])
```

```python
data = pd.concat([df[['Id','SalePrice']].reset_index(drop=True), pd.DataFrame(scaler.transform(df[feature_scale]), columns=feature_scale)], axis=1)
```
