# EDA ( Data Analysis)

```python
# import all the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# enable pandas to show all the columns of the dataset
pd.pandas.set_option('display.max_columns',None)
```

```python
# read the dataset 
dataset = pd.read_csv('Dataset/train.csv')
print(dataset.shape)
dataset.head()
```

# Types of features:
1. Nan Value features.
2. Numerical features:
    - Discrete features
    - Temporal features
    - Continuous features
3. Categorical features    

## 1. NaN values

```python
features_with_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum() > 1]
print('Total Number of features with NaN values: ',len(features_with_nan), '\n')

nan_percent = [np.round(dataset[feature].isnull().mean(), 4) for feature in features_with_nan]

plt.figure(figsize=(15,8))
sns.barplot(features_with_nan, nan_percent)
plt.title('Number of NaN values | Before')
plt.xticks(rotation=70, fontsize=11)
plt.ylabel('Count Percentage', fontsize=12)
plt.grid(1)
```

```python
for feature in features_with_nan:
    print(feature, np.round(dataset[feature].isnull().mean(), 4),  ' % missing values')
```

### since there are many missing values in the dataset, so we need to find the relationship between Missing values and target feature --> Sale Price
Lets plot some diagrams

```python


for feature in features_with_nan:
    data = dataset.copy()
    
    # indicates 1 when there is missing value other wise indicates 0
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    # Let's calculate the median of Sale price where the info is missing
    data.groupby(feature)['SalePrice'].median().plot.bar(colormap='Accent') 
    plt.title(feature)
    plt.ylabel('SalePrice')
    plt.xlabel(None)
    plt.show()
```

##### Here with the relation between NaN values and Dependent variables, we can clearly see that there are many missing values in the data. So we need to replace these NaN values to something meaningful.
But first we need to explore other type of features.

## 2. Numerical Features

```python
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O'] # O:Object
print('Total number of numerical features: ', len(numerical_features))
dataset[features_with_nan].head()
```

We got all Numerical features stored in the variable 'numerical_features', but there are <strong>Temporal Features</strong> also present like Date/Time/year.

## 2.1 Temporal Features

```python
temporal_features = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
print('Total number of Temporal features: ', len(temporal_features), '\n')

temporal_features

for feature in temporal_features:
    print(feature,'\n' ,dataset[feature].unique(), '\n')
```

#### Now let's find the relationship between <em>temporal_features</em> and <em>SalePrice</em>

```python
data = dataset.copy()

data.groupby(temporal_features[-1])['SalePrice'].median().plot()
plt.ylabel('Median House Price')
plt.grid(1)
```

#### Here we see a downward trend of Median House Prices with increase in time, which is not acceptable. Something is wrong here.
##### Now we will all other <strong>temporal_features</strong> vs <strong>'SalePrice'</strong> and see if we get something useful or not.

```python
for feature in  temporal_features:
    if feature != 'YrSold':
        data= dataset.copy()
        # we will capture the difference between year variable and year the house was sold
        data[feature] = data['YrSold'] - data[feature]
        
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

<em> Now it made sense! </em>

What did we observe from this?

<strong>YearBuilt</strong> : The older the house got, the cheaper it was sold.
            
<strong>YearRemodAdd</strong> : The more the time passed after the house was Remodeled, the cheaper it was sold.
            
<strong>GarageYrBlt</strong> : The older the garage, the lesser the price.
            
Basically, it's obvious that if the house is old, it'll be sold at a lower price than the house which is new.

Now, let's find out the number of <strong>Discrete Features</strong> in out dataset.

## 2.2 Discrete Features

```python
discrete_features = [feature for feature in numerical_features if len(dataset[feature].unique()) < 25 and feature not in temporal_features + ['Id']]
print('Total number of discrete features: ', len(discrete_features),'\n')
discrete_features
```

```python
dataset[discrete_features].head()
```

Now lets find the relation between <strong>discrete_features</strong> and <strong>SalePrice(target feature)</strong>

```python
for feature in discrete_features:
    data = dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.show()
```

# 2.3 Continous Features

```python
continuous_features = [feature for feature in numerical_features if feature not in discrete_features + temporal_features +['Id']]
print('Total number of Continuous features: ', len(continuous_features), '\n')
continuous_features
```

```python
dataset[continuous_features].head()
```

```python
for feature in continuous_features:
    data = dataset.copy()
    data[feature].hist(bins=20)
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.show()
```

We notice that we cannot gather much information about these features because this is a not a Normally distributed dataset (+skewed) with less or no information. Thus, we'll try to <strong>log normalize</strong>.

# 2.3.1. Applying Log Normal Distribution to the Continuous Features

```python
for feature in continuous_features:
    data= dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] =np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

#### Now we get an evenly spread graph to perform analysis on: 

-We can see that the plot are showing us a monotonix behaviour (i.e, when the feature is increasing, the SalePrice is also increasing.)

-It is also showing us a positive co-relation

## 2.3.2 Outliers
With the help of BoxPlot, we can see the outliers of the continuous features ONLY.

```python

for feature in continuous_features:
    data= dataset.copy()
    
    if 0 not in data[feature].unique():
        data[feature] = np.log(data[feature])
        data.boxplot(feature)
        plt.title(feature)
        plt.grid(1)
        plt.show()
        
```

##### As we can see, there are a lot of multipliers

## 3. Categorical features

```python
categorical_features = [feature for feature in dataset.columns if data[feature].dtypes == 'O']
print('Total number of categorical features: ', len(categorical_features), '\n')
categorical_features
```

```python
dataset[categorical_features].head()
```

##### First, we'll find the Cardinal Values for the features. It seems high-level stuff but it is simply the count of Categories inside each of the categorical feature ;p

```python
for feature in categorical_features:
    print('The feature is {} and number of categories are {}'.format(feature,len(dataset[feature].unique())))
```

```python
categories = []
for f in categorical_features:
#     print(f'{f} > {len(dataset[f].unique())}')
    categories.append(len(dataset[f].unique()))

plt.figure(figsize=(20, 10))
sns.barplot(categorical_features, categories)

plt.yticks(range(0, 26, 1))
plt.xticks(rotation=80, fontsize=18)
plt.ylabel('Count', fontsize=18)
plt.grid(1)
```

## Categorical features v/s Sale price

```python
for feature in categorical_features:
    data = dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature, fontsize=12)
    plt.ylabel('SalePrice', fontsize=12)
    plt.xticks(rotation=80)
    plt.show()
```
