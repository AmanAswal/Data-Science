```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('dataset/eda.csv')
print(df.shape)
df.head()
```

# 1. Wrangle Data

#### We see that, Sex and Embarked have only very few classes, thus, we'll one-hot encode them.

```python
sex = pd.get_dummies(df['Sex'], drop_first=True)             # We can drop_first because it'll behave just like binary code.
embarked = pd.get_dummies(df['Embarked'], drop_first=True)

# Now, we'll add these features to our dataset.
df = pd.concat([df, sex, embarked], axis=1)

df.head()    
```

### We will drop the columns now

```python
df.drop(['PassengerId', 'Name', 'Embarked', 'Ticket', 'Sex'], axis=1, inplace=True)
```

```python
df.head()
```

## Now our dataset is ready.

```python
df.to_csv('dataset/final.csv', index=False)
```
